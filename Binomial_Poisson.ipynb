# Poisson Distribution

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Parameters used
p_defective = 0.001  # Probability of a bottle being defective
bottles_per_box = 500  # Number of bottles in a box
boxes_purchased = 100  # Number of boxes purchased
lambda_val = bottles_per_box * p_defective
```

```python
#function to calculate probabilities using the Poisson distribution
def calculate_probabilities(lmbda):
    # Probability of no defective bottles (k=0)
    prob_no_defective = poisson.pmf(0, lmbda)

    # Probability of at least 2 defective bottles
    prob_at_least_2 = 1 - (poisson.pmf(0, lmbda) + poisson.pmf(1, lmbda))

    # Probability of at most 2 defective bottles
    prob_at_most_2 = poisson.cdf(2, lmbda)

    return prob_no_defective, prob_at_least_2, prob_at_most_2
```

```python
# Calculating probabilities
prob_no_def, prob_at_least_2, prob_at_most_2 = calculate_probabilities(lambda_val)

# Calculating the expected number of boxes satisfying each condition
boxes_no_def = boxes_purchased * prob_no_def
boxes_at_least_2 = boxes_purchased * prob_at_least_2
boxes_at_most_2 = boxes_purchased * prob_at_most_2
```

```python
# Visualization 1: Probability of each condition
conditions = ["No Defective", "At Least 2 Defective", "At Most 2 Defective"]
probabilities = [prob_no_def, prob_at_least_2, prob_at_most_2]

plt.figure(figsize=(10, 5))
plt.bar(conditions, probabilities, color=['blue', 'orange', 'green'])
plt.title("Probability of Different Defective Conditions")
plt.ylabel("Probability")
plt.xlabel("Conditions")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
# Visualization 2: Expected number of boxes for each condition
boxes_count = [boxes_no_def, boxes_at_least_2, boxes_at_most_2]

plt.figure(figsize=(10, 5))
plt.bar(conditions, boxes_count, color=['blue', 'orange', 'green'])
plt.title("Expected Number of Boxes Satisfying Each Condition")
plt.ylabel("Number of Boxes")
plt.xlabel("Conditions")
plt.ylim(0, max(boxes_count) + 5)
plt.grid(axis='y')
plt.show()
```

```python
# Print the results
print("Results:")
print(f"Probability of no defective bottles: {prob_no_def:.4f}")
print(f"Probability of at least 2 defective bottles: {prob_at_least_2:.4f}")
print(f"Probability of at most 2 defective bottles: {prob_at_most_2:.4f}")
print(f"Expected number of boxes with no defective bottles: {boxes_no_def:.2f}")
print(f"Expected number of boxes with at least 2 defective bottles: {boxes_at_least_2:.2f}")
print(f"Expected number of boxes with at most 2 defective bottles: {boxes_at_most_2:.2f}")
```

# Binomial Distribution

```python
# Importing necessary libraries
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# Parameters used
n = 10  # Number of coin tosses
p_head = 0.5  # Probability of getting a head
```

```python
def calculate_binomial_probabilities(n, p):
    # Probability of getting exactly 3 heads
    prob_exactly_3 = binom.pmf(3, n, p)

    # Probability of getting at least 3 heads (1 - P(X <= 2))
    prob_at_least_3 = 1 - binom.cdf(2, n, p)

    # Probability of getting at most 3 heads (P(X <= 3))
    prob_at_most_3 = binom.cdf(3, n, p)

    # Probability of getting at least 7 heads (1 - P(X <= 6))
    prob_at_least_7 = 1 - binom.cdf(6, n, p)

    return prob_exactly_3, prob_at_least_3, prob_at_most_3, prob_at_least_7
```

```python
 #Calculating the probabilities
prob_exactly_3, prob_at_least_3, prob_at_most_3, prob_at_least_7 = calculate_binomial_probabilities(n, p_head)
```

```python
# Visualizing the probabilities for better understanding
conditions = ["Exactly 3 Heads", "At Least 3 Heads", "At Most 3 Heads", "At Least 7 Heads"]
probabilities = [prob_exactly_3, prob_at_least_3, prob_at_most_3, prob_at_least_7]
```

```python
# Ploting the probabilities
plt.figure(figsize=(10, 5))
plt.bar(conditions, probabilities, color=['blue', 'orange', 'green', 'purple'])
plt.title("Binomial Probabilities for Coin Toss Experiment")
plt.ylabel("Probability")
plt.xlabel("Conditions")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
#results
print("Results:")
print(f"Probability of exactly 3 heads: {prob_exactly_3:.4f}")
print(f"Probability of at least 3 heads: {prob_at_least_3:.4f}")
print(f"Probability of at most 3 heads: {prob_at_most_3:.4f}")
print(f"Probability of at least 7 heads: {prob_at_least_7:.4f}")
```
